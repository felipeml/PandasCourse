# Exercicio boxplot

```python
dados = pd.read_csv('dados/aluguel_amostra.csv', sep = ';')
```

```python
dados
```

```python
valor_m2 = dados['Valor m2']
```

```python
valor_m2
```

```python
q1 = valor_m2.quantile(.25)
q3 = valor_m2.quantile(.75)
iiq = (q3 - q1).round(2)
limite_inferior = (q1 - 1.5 * iiq).round(2)
limite_superior = q3 + 1.5 * iiq
```

```python
print('[Q1] -> {}'.format(q1))
print('[Q3] -> {}'.format(q3))
print('[IIQ] -> {}'.format(iiq))
print('[Q1 - 1.5 * IIQ] -> {}'.format(limite_inferior))
print('[Q3 + 1.5 * IIQ] -> {}'.format(limite_superior))
```
